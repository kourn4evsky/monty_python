# Задание 1
Дан текст:
*"варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг"*
+ Необходимо выделить все биграммы в этом тексте и сохранить их список в какую-либо переменную. 

```python
text = "варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг"
words = text.split()
bgrams = []
for i in range(28): # В этой строчке прописано максимальное колличество комбинаций, чтобы избежать ошибки, когда закончатся биграммы.
     bgram = words[i] + " " + words[i+1] 
     bgrams.append(bgram) # Теперь каждая новая биграмма будет автоматически добавляться в список.
print(bgrams)
print(len(bgrams))
```
